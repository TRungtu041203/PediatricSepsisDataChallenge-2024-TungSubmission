```python
# ====================================================================
# OPTIMIZED PREPROCESSING WITH CACHING
# ====================================================================

print("ğŸ”§ PREPROCESSING WITH CACHING")
print("="*40)

target = "inhospital_mortality"

# Check for cached preprocessing
cache_file = "preprocessing_pipeline.pkl"
try:
    print("ğŸ” Checking for cached preprocessing...")
    with open(cache_file, 'rb') as f:
        cached_data = pickle.load(f)
    
    preproc = cached_data['preprocessor']
    X_train_processed = cached_data['X_train_processed']
    X_test_processed = cached_data['X_test_processed']
    y_train = cached_data['y_train']
    y_test = cached_data['y_test']
    
    print("âœ… Loaded cached preprocessing!")
    print(f"  â€¢ Train matrix shape: {X_train_processed.shape}")
    print(f"  â€¢ Test matrix shape: {X_test_processed.shape}")
    
except FileNotFoundError:
    print("ğŸ“ No cache found - building preprocessing pipeline...")
    
    # Create missing value flags
    num_cols_raw = (
        df.select_dtypes(include=["float64", "int64", "Int8"])
          .columns.drop(target)
          .tolist()
    )
    manual_flags = [c for c in df.columns if c.endswith("_missing")]
    cols_needing_flag = [c for c in num_cols_raw if f"{c}_missing" not in manual_flags]

    for col in cols_needing_flag:
        df[f"{col}_missing"] = df[col].isna().astype("Int8")

    # Recompute column groups after the new flags were added
    num_cols = (
        df.select_dtypes(include=["float64", "int64", "Int8"])
          .columns.drop(target)
          .tolist()
    )
    cat_cols = (
        df.select_dtypes(include="category")
          .columns.drop(target, errors="ignore")
          .tolist()
    )

    print(f"â„¹ï¸  Numeric cols: {len(num_cols)}")
    print(f"â„¹ï¸  Categorical cols: {len(cat_cols)}")
    print(f"â„¹ï¸  Missing flags: {len([c for c in num_cols if c.endswith('_missing')])}")

    # Create preprocessing pipelines
    numeric_pipe = Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scale",   RobustScaler(quantile_range=(10, 90))),
    ])

    categorical_pipe = Pipeline([
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("encoder", OneHotEncoder(
            handle_unknown="ignore",
            sparse_output=False,
            drop="first"
        )),
    ])

    preproc = ColumnTransformer([
        ("num", numeric_pipe, num_cols),
        ("cat", categorical_pipe, cat_cols),
    ])

    # Train/test split
    X = df.drop(columns=[target])
    y = df[target]

    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.20,
        random_state=42,
        stratify=y,
    )

    # Fit and transform
    print("ğŸ”„ Fitting preprocessor...")
    X_train_processed = preproc.fit_transform(X_train)
    X_test_processed = preproc.transform(X_test)

    print(f"âœ… Train matrix shape: {X_train_processed.shape}")
    print(f"âœ… Test matrix shape: {X_test_processed.shape}")
    
    # Cache the results
    print("ğŸ’¾ Caching preprocessing results...")
    cache_data = {
        'preprocessor': preproc,
        'X_train_processed': X_train_processed,
        'X_test_processed': X_test_processed,
        'y_train': y_train,
        'y_test': y_test
    }
    
    with open(cache_file, 'wb') as f:
        pickle.dump(cache_data, f)
    print(f"âœ… Cached preprocessing to {cache_file}")

print(f"\nğŸ“Š Dataset Summary:")
print(f"  â€¢ Total samples: {len(y_train) + len(y_test)}")
print(f"  â€¢ Train: {len(y_train)} samples")
print(f"  â€¢ Test: {len(y_test)} samples")
print(f"  â€¢ Class 0 (survived): {sum(y_train == 0)} ({sum(y_train == 0)/len(y_train)*100:.1f}%)")
print(f"  â€¢ Class 1 (died): {sum(y_train == 1)} ({sum(y_train == 1)/len(y_train)*100:.1f}%)")
print(f"  â€¢ Imbalance ratio: {sum(y_train == 0)/sum(y_train == 1):.1f}:1")
print("")
```

```python
# ====================================================================
# ENHANCED FEATURE SELECTION WITH ENGINEERED FEATURE PROTECTION
# ====================================================================

print("ğŸ¯ ENHANCED FEATURE SELECTION - PROTECTING ENGINEERED FEATURES")
print("="*65)
print("âœ… Clinical domain knowledge + Engineered feature protection")
print("âœ… Conservative selection to preserve clinical insights")
print("âœ… Prioritize newly created features from preprocessing")
print("")

from sklearn.feature_selection import (
    VarianceThreshold, SelectKBest, f_classif, mutual_info_classif,
    RFE, SelectFromModel
)
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import chi2_contingency
import warnings
warnings.filterwarnings('ignore')

# ====================================================================
# 0. IDENTIFY AND CATEGORIZE ALL FEATURES
# ====================================================================

print("ğŸ” STEP 0: Feature Categorization & Protection Setup")
print("-" * 55)

# Get feature names from preprocessor
if hasattr(preproc, 'get_feature_names_out'):
    all_feature_names = list(preproc.get_feature_names_out())
else:
    all_feature_names = [f"feature_{i}" for i in range(X_train_processed.shape[1])]

print(f"ğŸ“Š Total features in dataset: {len(all_feature_names)}")

# 1. Core Clinical Features (Always Protected)
essential_clinical_patterns = [
    'agecalc_adm', 'weight_kg_adm', 'hr_bpm_adm', 'sysbp_mmhg_adm', 
    'diasbp_mmhg_adm', 'rr_brpm_app_adm', 'temp_c_adm', 'spo2',
    'lactate_mmolpl_adm', 'glucose_mmolpl_adm', 'hematocrit_gpdl_adm'
]

# 2. Engineered Features (High Priority Protection) - CORRECTED FROM PREPROCESSING
engineered_patterns = [
    'map_mmhg',           # Mean Arterial Pressure = (SBP + 2Ã—DBP)/3
    'shock_index',        # Shock Index = HR/SBP
    'hr_pct_of_limit',   # Heart rate percentage of limit
    'tachy_flag',        # Tachycardia flag
    'on_oxygen_flag',    # Oxygen availability flag
    'hypoxia_flag',      # Hypoxia flag
    'bcs_total',         # Blantyre Coma Scale total
    'coma_flag',         # Coma flag
    'lactate_cat',       # Lactate categories
    'map_mmhg_missing',  # MAP missing flag
    'shock_index_missing', # Shock index missing flag
    'hr_pct_of_limit_missing', # HR limit missing flag
    'lactate_missing',   # Lactate missing flag
    'bcsmotor_adm_missing', # BCS motor missing
    'bcseye_adm_missing',   # BCS eye missing
    'bcsverbal_adm_missing' # BCS verbal missing
]

# 3. Missing Value Indicators (Medium Priority)
missing_patterns = ['_missing']

# Categorize features
core_clinical = []
engineered_features = []
missing_indicators = []
other_features = []

for feature in all_feature_names:
    feature_lower = feature.lower()
    
    # Check if it's a core clinical feature
    if any(pattern in feature_lower for pattern in essential_clinical_patterns):
        core_clinical.append(feature)
    # Check if it's an engineered feature
    elif any(pattern in feature_lower for pattern in engineered_patterns):
        engineered_features.append(feature)
    # Check if it's a missing indicator
    elif any(pattern in feature_lower for pattern in missing_patterns):
        missing_indicators.append(feature)
    else:
        other_features.append(feature)

print(f"ğŸ¥ Core clinical features: {len(core_clinical)}")
print(f"ğŸ”§ Engineered features: {len(engineered_features)}")
print(f"ğŸ“Š Missing indicators: {len(missing_indicators)}")
print(f"ğŸ“‹ Other features: {len(other_features)}")

# Create protection levels
high_priority = core_clinical + engineered_features  # Never remove
medium_priority = missing_indicators                  # Remove only if necessary
low_priority = other_features                        # Can remove freely

print(f"ğŸ›¡ï¸  High priority (protected): {len(high_priority)}")
print(f"âš–ï¸  Medium priority: {len(medium_priority)}")
print(f"ğŸ“ Low priority: {len(low_priority)}")

print(f"\nğŸ”§ Sample engineered features found: {engineered_features[:5]}")
print(f"ğŸ¥ Sample core clinical features: {core_clinical[:5]}")

# ====================================================================
# 1. CONSERVATIVE VARIANCE FILTERING
# ====================================================================

print(f"\nğŸ” STEP 1: Conservative Variance Filtering (Protect Engineered)")
print("-" * 60)

# Use very low threshold to only remove truly constant features
variance_selector = VarianceThreshold(threshold=0.0001)  # Almost zero variance only
X_variance_filtered = variance_selector.fit_transform(X_train_processed)

# Get surviving features
variance_mask = variance_selector.get_support()
surviving_features = [all_feature_names[i] for i in range(len(all_feature_names)) if variance_mask[i]]

# Check if any protected features were removed
removed_protected = [f for f in high_priority if f not in surviving_features]
if removed_protected:
    print(f"âš ï¸  WARNING: {len(removed_protected)} protected features removed by variance filter!")
    print(f"ğŸ”§ Consider adjusting variance threshold")

print(f"ğŸ“Š Before variance filtering: {len(all_feature_names)} features")
print(f"ğŸ“Š After variance filtering: {len(surviving_features)} features")
print(f"âŒ Removed: {len(all_feature_names) - len(surviving_features)} near-constant features")

# ====================================================================
# 2. ENGINEERED-AWARE CORRELATION FILTERING
# ====================================================================

print(f"\nğŸ“ˆ STEP 2: Correlation Analysis (Prioritize Engineered Features)")
print("-" * 65)

# Create correlation matrix
temp_df = pd.DataFrame(X_variance_filtered, columns=surviving_features)
correlation_matrix = temp_df.corr().abs()

# Find highly correlated pairs
high_corr_pairs = []
correlation_threshold = 0.95

for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if correlation_matrix.iloc[i, j] > correlation_threshold:
            feat1, feat2 = correlation_matrix.columns[i], correlation_matrix.columns[j]
            high_corr_pairs.append((feat1, feat2, correlation_matrix.iloc[i, j]))

print(f"ğŸ” Found {len(high_corr_pairs)} highly correlated pairs (r > {correlation_threshold})")

# Smart removal prioritizing engineered features
features_to_remove = set()
engineered_protected = 0

for feat1, feat2, corr_val in high_corr_pairs:
    # Check feature priorities
    feat1_is_high_priority = feat1 in high_priority
    feat2_is_high_priority = feat2 in high_priority
    
    # Get target correlations
    feat1_idx = surviving_features.index(feat1)
    feat2_idx = surviving_features.index(feat2)
    
    try:
        feat1_target_corr = abs(np.corrcoef(X_variance_filtered[:, feat1_idx], y_train)[0, 1])
        feat2_target_corr = abs(np.corrcoef(X_variance_filtered[:, feat2_idx], y_train)[0, 1])
    except:
        feat1_target_corr, feat2_target_corr = 0, 0
    
    # Decision logic prioritizing engineered features
    if feat1_is_high_priority and not feat2_is_high_priority:
        features_to_remove.add(feat2)
        if feat1 in engineered_features:
            engineered_protected += 1
    elif feat2_is_high_priority and not feat1_is_high_priority:
        features_to_remove.add(feat1)
        if feat2 in engineered_features:
            engineered_protected += 1
    elif feat1_is_high_priority and feat2_is_high_priority:
        # Both are high priority - keep the one with higher target correlation
        if feat1_target_corr < feat2_target_corr:
            features_to_remove.add(feat1)
        else:
            features_to_remove.add(feat2)
    else:
        # Neither is high priority - use target correlation
        if feat1_target_corr < feat2_target_corr:
            features_to_remove.add(feat1)
        else:
            features_to_remove.add(feat2)

# Apply correlation filtering
correlation_filtered_features = [f for f in surviving_features if f not in features_to_remove]
correlation_filtered_indices = [surviving_features.index(f) for f in correlation_filtered_features]
X_correlation_filtered = X_variance_filtered[:, correlation_filtered_indices]

print(f"ğŸ”§ Protected {engineered_protected} engineered features from correlation removal")
print(f"âŒ Removed {len(features_to_remove)} highly correlated features")
print(f"ğŸ“Š Remaining after correlation filtering: {len(correlation_filtered_features)} features")

# ====================================================================
# 3. CONSERVATIVE STATISTICAL SELECTION
# ====================================================================

print(f"\nğŸ“Š STEP 3: Statistical Selection (Keep Most Features)")
print("-" * 55)

# Keep 90% of features to be very conservative
keep_ratio = 0.9
target_features = max(len(high_priority), int(len(correlation_filtered_features) * keep_ratio))

if target_features >= len(correlation_filtered_features):
    # Keep all features
    X_statistical_filtered = X_correlation_filtered
    statistical_features = correlation_filtered_features
    print(f"ğŸ“Š Keeping all {len(correlation_filtered_features)} features (conservative approach)")
else:
    # Use statistical selection
    mi_selector = SelectKBest(score_func=mutual_info_classif, k=target_features)
    X_statistical_filtered = mi_selector.fit_transform(X_correlation_filtered, y_train)
    
    statistical_mask = mi_selector.get_support()
    statistical_features = [correlation_filtered_features[i] for i in range(len(correlation_filtered_features)) if statistical_mask[i]]
    
    print(f"ğŸ“Š Selected top {len(statistical_features)} features by mutual information")

# Check protected features
remaining_protected = [f for f in high_priority if f in statistical_features]
print(f"ğŸ›¡ï¸  Protected features retained: {len(remaining_protected)}/{len(high_priority)}")

# ====================================================================
# 4. FINAL FEATURE SET WITH MANDATORY INCLUSIONS
# ====================================================================

print(f"\nğŸ¯ STEP 4: Final Feature Set Assembly")
print("-" * 45)

# Ensure all high priority features are included
missing_protected = [f for f in high_priority if f not in statistical_features]

if missing_protected:
    print(f"ğŸ”§ Adding back {len(missing_protected)} missing protected features")
    
    # Find indices of missing protected features in original data
    missing_indices = []
    for feature in missing_protected:
        if feature in all_feature_names:
            original_idx = all_feature_names.index(feature)
            missing_indices.append(original_idx)
    
    if missing_indices:
        missing_data = X_train_processed[:, missing_indices]
        X_final = np.concatenate([X_statistical_filtered, missing_data], axis=1)
        final_features = statistical_features + missing_protected
    else:
        X_final = X_statistical_filtered
        final_features = statistical_features
else:
    X_final = X_statistical_filtered
    final_features = statistical_features

print(f"âœ… Final feature set assembled: {len(final_features)} features")

# ====================================================================
# 5. FEATURE PROTECTION SUMMARY
# ====================================================================

print(f"\nğŸ“‹ ENHANCED FEATURE SELECTION SUMMARY")
print("="*50)

# Count feature types in final set
final_core = [f for f in final_features if f in core_clinical]
final_engineered = [f for f in final_features if f in engineered_features]
final_missing = [f for f in final_features if f in missing_indicators]
final_other = [f for f in final_features if f in other_features]

original_count = len(all_feature_names)
reduction_ratio = (1 - len(final_features)/original_count) * 100

print(f"ğŸ“Š FEATURE PRESERVATION RESULTS:")
print(f"  â€¢ Original features:        {original_count}")
print(f"  â€¢ Final features:           {len(final_features)} ({(len(final_features)/original_count)*100:.1f}% kept)")
print(f"  â€¢ Reduction ratio:          {reduction_ratio:.1f}% removed")
print(f"")
print(f"ğŸ¥ Core clinical retained:    {len(final_core)}/{len(core_clinical)} ({len(final_core)/len(core_clinical)*100:.1f}%)")
print(f"ğŸ”§ Engineered retained:       {len(final_engineered)}/{len(engineered_features)} ({len(final_engineered)/max(1,len(engineered_features))*100:.1f}%)")
print(f"ğŸ“Š Missing indicators:        {len(final_missing)}/{len(missing_indicators)} ({len(final_missing)/max(1,len(missing_indicators))*100:.1f}%)")
print(f"ğŸ“‹ Other features:            {len(final_other)}/{len(other_features)} ({len(final_other)/max(1,len(other_features))*100:.1f}%)")

# Assessment
if reduction_ratio < 20:
    assessment = "âœ… VERY CONSERVATIVE - Excellent feature preservation"
elif reduction_ratio < 40:
    assessment = "âœ… CONSERVATIVE - Good feature preservation"
elif reduction_ratio < 60:
    assessment = "âš–ï¸  BALANCED - Reasonable feature reduction"
else:
    assessment = "âš ï¸  AGGRESSIVE - Significant feature reduction"

print(f"  â€¢ Selection Assessment:     {assessment}")

print(f"\nğŸ”§ ENGINEERED FEATURES IN FINAL SET:")
print("-" * 40)
for i, feature in enumerate(final_engineered, 1):
    clean_name = feature.replace('num__', '').replace('cat__', '')
    print(f"{i:2d}. {clean_name}")

if len(final_engineered) == 0:
    print("âš ï¸  No engineered features found - check feature naming patterns")

print(f"\nğŸ¥ CORE CLINICAL FEATURES IN FINAL SET:")
print("-" * 40)
for i, feature in enumerate(final_core[:10], 1):
    clean_name = feature.replace('num__', '').replace('cat__', '')
    print(f"{i:2d}. {clean_name}")

if len(final_core) > 10:
    print(f"    ... and {len(final_core) - 10} more clinical features")

# Apply transformations to test set
print(f"\nğŸ”„ Applying feature selection to test set...")

X_test_variance = variance_selector.transform(X_test_processed)
X_test_correlation = X_test_variance[:, correlation_filtered_indices]

if target_features < len(correlation_filtered_features) and 'mi_selector' in locals():
    X_test_statistical = mi_selector.transform(X_test_correlation)
else:
    X_test_statistical = X_test_correlation

if missing_protected and missing_indices:
    missing_test_data = X_test_processed[:, missing_indices]
    X_test_final = np.concatenate([X_test_statistical, missing_test_data], axis=1)
else:
    X_test_final = X_test_statistical

# Update global variables
X_train_processed = X_final
X_test_processed = X_test_final
feature_names = final_features

print(f"âœ… Enhanced feature selection complete!")
print(f"ğŸ“‹ X_train_processed: {X_train_processed.shape}")
print(f"ğŸ“‹ X_test_processed: {X_test_processed.shape}")
print(f"ğŸ“‹ feature_names: {len(feature_names)} features")

# Store enhanced results
enhanced_feature_results = {
    'original_features': original_count,
    'final_features': len(final_features),
    'reduction_ratio': reduction_ratio / 100,
    'selected_features': final_features,
    'feature_categories': {
        'core_clinical': final_core,
        'engineered': final_engineered,
        'missing_indicators': final_missing,
        'other': final_other
    },
    'protection_stats': {
        'core_clinical_retention': len(final_core) / max(1, len(core_clinical)),
        'engineered_retention': len(final_engineered) / max(1, len(engineered_features)),
        'missing_indicators_retention': len(final_missing) / max(1, len(missing_indicators))
    },
    'selection_approach': 'engineered_feature_protection',
    'thresholds': {
        'variance_threshold': 0.0001,
        'correlation_threshold': 0.95,
        'statistical_keep_ratio': keep_ratio
    }
}

globals()['enhanced_feature_results'] = enhanced_feature_results
print("ğŸ“Š Enhanced feature selection results stored in 'enhanced_feature_results'")
```

```python
# ====================================================================
# OPTIMIZED SAMPLING STRATEGIES WITH MINORITY-ONLY SMOTE
# ====================================================================

print("ğŸ¯ OPTIMIZED SAMPLING STRATEGIES")
print("="*40)
print("âœ… SMOTE with sampling_strategy='auto' for minority only")
print("âœ… Fixed k_neighbors for small minority class")
print("")

# Create optimized samplers
def get_optimized_ros():
    """RandomOverSampler with auto balancing"""
    return RandomOverSampler(
        random_state=42,
        sampling_strategy='auto'  # Balance to majority class size
    )

def get_optimized_smote():
    """SMOTE with minority-only balancing"""
    return SMOTE(
        random_state=42,
        k_neighbors=min(3, sum(y_train == 1) - 1),  # Adapt to minority class size
        sampling_strategy='auto'  # Only balance minority class
    )

def get_optimized_kmeans_smote():
    """KMeans-SMOTE with fallback to SMOTE"""
    try:
        return KMeansSMOTE(
            random_state=42,
            k_neighbors=min(2, sum(y_train == 1) - 1),
            cluster_balance_threshold=0.01,
            sampling_strategy='auto'
        )
    except:
        # Fallback to regular SMOTE
        return SMOTE(
            random_state=42,
            k_neighbors=min(2, sum(y_train == 1) - 1),
            sampling_strategy='auto'
        )

def get_optimized_smote_enn():
    """SMOTE-ENN with optimized parameters"""
    try:
        return SMOTEENN(
            random_state=42,
            smote=SMOTE(
                random_state=42, 
                k_neighbors=min(2, sum(y_train == 1) - 1),
                sampling_strategy='auto'
            )
        )
    except:
        # Fallback to regular SMOTE
        return SMOTE(
            random_state=42,
            k_neighbors=min(2, sum(y_train == 1) - 1),
            sampling_strategy='auto'
        )

sampling_methods = {
    'ROS': get_optimized_ros(),
    'SMOTE': get_optimized_smote(),
    'KMeans-SMOTE': get_optimized_kmeans_smote(),
    'SMOTE-ENN': get_optimized_smote_enn()
}

print("âœ… Created optimized sampling strategies:")
for name, sampler in sampling_methods.items():
    print(f"  â€¢ {name}: {type(sampler).__name__}")
print("")
```

```python
# ====================================================================
# OPTIMISED MODELS WITH IMPROVED BASELINES
# ====================================================================

print("ğŸ”„  CREATING OPTIMISED MODELS & SAMPLING STRATEGIES")
print("===================================================")
print("âœ…  BalancedRandomForest & EasyEnsemble baselines")
print("âœ…  SMOTE/ROS will use sampling_strategy='auto'")
print("âœ…  Solver/penalty combos fixed for Logistic R.")
print("âœ…  XGBoost optimised for AUPRC\n")

# --------------------------------------------------------------------
# helper constructors
# --------------------------------------------------------------------
def get_logistic_regression():
    """Elastic-net ready LogisticRegression (only solvers that support it)."""
    return LogisticRegression(
        solver="saga",                  # supports l1/l2/elastic-net
        penalty="elasticnet",
        l1_ratio=0.5,
        max_iter=2000,
        class_weight="balanced",
        random_state=42,
    )

def get_random_forest():
    """Standard RandomForest with class balance."""
    return RandomForestClassifier(
        n_estimators=300,
        max_depth=None,
        min_samples_leaf=1,
        class_weight="balanced",
        n_jobs=-1,
        random_state=42,
    )

def get_xgboost_auprc():
    """XGBoost optimised for AUPRC."""
    pos_weight = len(y_train[y_train == 0]) / len(y_train[y_train == 1])
    return XGBClassifier(
        n_estimators=300,
        max_depth=4,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        tree_method="hist",
        objective="binary:logistic",
        eval_metric="aucpr",
        scale_pos_weight=pos_weight,
        n_jobs=-1,
        random_state=42,
    )

def get_balanced_random_forest():
    """BalancedRandomForest â€“ native under-sampling inside each tree."""
    return BalancedRandomForestClassifier(
        n_estimators=300,
        max_depth=None,
        min_samples_leaf=1,
        n_jobs=-1,
        random_state=42,
        sampling_strategy="auto",
    )

def get_easy_ensemble():
    """EasyEnsemble â€“ bag of balanced AdaBoost trees."""
    return EasyEnsembleClassifier(
        n_estimators=10,           # 10 balanced subsets Ã— 50 trees each (default)
        sampling_strategy="auto",
        n_jobs=-1,
        random_state=42,
    )

# --------------------------------------------------------------------
# dictionary of ready-to-use base models
# --------------------------------------------------------------------
base_models = {
    "Logistic Regression"     : get_logistic_regression(),
    "Random Forest"           : get_random_forest(),
    "XGBoost"                 : get_xgboost_auprc(),
    "Balanced Random Forest"  : get_balanced_random_forest(),
    "Easy Ensemble"           : get_easy_ensemble(),
}

print("âœ…  Optimised base models created:")
for name, model in base_models.items():
    print(f"   â€¢ {name:24s} â†’ {type(model).__name__}")
print()
```

```python
# ====================================================================
# OPTIMIZED SEARCH SPACES FOR BAYESIAN/RANDOM SEARCH
# ====================================================================

print("ğŸ” BAYESIAN/RANDOM SEARCH OPTIMIZATION")
print("="*45)
print("âœ… Replaced exhaustive GridSearch with efficient Bayesian/Random search")
print("âœ… Fixed solver/penalty incompatible combinations")
print("âœ… Reduced search space while maintaining coverage")
print("")

# Define compatible solver/penalty combinations for LogisticRegression
def get_logistic_params(is_pipeline=True):
    """Get compatible LogisticRegression parameter distributions"""
    prefix = 'classifier__' if is_pipeline else ''
    return {
        f'{prefix}C': Real(0.01, 100, prior='log-uniform'),
        f'{prefix}max_iter': Integer(1000, 3000),
        # Only use compatible solver/penalty combinations
        f'{prefix}solver': Categorical(['liblinear', 'saga']),
        f'{prefix}penalty': Categorical(['l1', 'l2']),  # elasticnet needs saga only
        f'{prefix}class_weight': Categorical(['balanced', None])
    }

def get_random_forest_params(is_pipeline=True):
    """RandomForest parameter distributions"""
    prefix = 'classifier__' if is_pipeline else ''
    return {
        f'{prefix}n_estimators': Integer(100, 500),
        f'{prefix}max_depth': Categorical([10, 15, 20, 25, None]),
        f'{prefix}min_samples_split': Integer(2, 20),
        f'{prefix}min_samples_leaf': Integer(1, 10),
        f'{prefix}bootstrap': Categorical([True, False]),
        f'{prefix}class_weight': Categorical(['balanced', None])
    }

def get_xgboost_params(is_pipeline=True):
    """XGBoost parameter distributions optimized for AUPRC"""
    prefix = 'classifier__' if is_pipeline else ''
    return {
        f'{prefix}n_estimators': Integer(100, 500),
        f'{prefix}max_depth': Integer(3, 8),
        f'{prefix}learning_rate': Real(0.01, 0.3, prior='log-uniform'),
        f'{prefix}subsample': Real(0.6, 1.0),
        f'{prefix}colsample_bytree': Real(0.6, 1.0),
        f'{prefix}gamma': Real(0, 5),
        f'{prefix}reg_alpha': Real(0, 1),
        f'{prefix}reg_lambda': Real(1, 2)
    }

# Sampling parameter distributions
def get_sampling_params(method):
    """Get sampling parameter distributions"""
    if method == 'ROS':
        return {'sampler__random_state': Categorical([42])}
    elif method == 'SMOTE':
        return {
            'sampler__random_state': Categorical([42]),
            'sampler__k_neighbors': Integer(1, min(5, sum(y_train == 1) - 1))
        }
    elif method == 'KMeans-SMOTE':
        return {
            'sampler__random_state': Categorical([42]),
            'sampler__k_neighbors': Integer(1, min(3, sum(y_train == 1) - 1)),
            'sampler__cluster_balance_threshold': Real(0.001, 0.1, prior='log-uniform')
        }
    elif method == 'SMOTE-ENN':
        return {
            'sampler__random_state': Categorical([42]),
            'sampler__smote__k_neighbors': Integer(1, min(3, sum(y_train == 1) - 1))
        }
    else:
        return {}

print("âœ… Created optimized parameter spaces:")
print("  â€¢ LogisticRegression: Compatible solver/penalty combinations")
print("  â€¢ RandomForest: Balanced parameter ranges")
print("  â€¢ XGBoost: AUPRC-optimized parameters")
print("  â€¢ Sampling: Adaptive k_neighbors based on minority class size")
print("")
```

```python
# ====================================================================
# OPTIMIZED PIPELINE CREATION AND SEARCH FUNCTIONS
# ====================================================================

print("ğŸ”§ PIPELINE CREATION & SEARCH FUNCTIONS")
print("="*45)

# Create optimized pipelines
def create_pipeline(model_name, sampler=None):
    """Create optimized pipeline with optional sampling"""
    model = base_models[model_name]
    
    if sampler is None:
        # No sampling - use processed data directly for balanced models
        return model
    else:
        # With sampling - use processed data with sampling
        return ImbPipeline([
            ('sampler', sampler),
            ('classifier', model)
        ])

def run_optimized_search(pipeline, param_space, pipeline_name, search_type='bayes', n_iter=50):
    """Run Bayesian or Random search with AUPRC optimization"""
    print(f"ğŸ” Optimizing: {pipeline_name}")
    
    # Define scoring - optimize for AUPRC
    scoring = {
        'auprc': make_scorer(average_precision_score, needs_proba=True),
        'roc_auc': make_scorer(roc_auc_score, needs_proba=True)
    }
    
    # CV strategy
    cv = RepeatedStratifiedKFold(n_splits=3, n_repeats=2, random_state=42)
    
    try:
        if search_type == 'bayes' and len(param_space) > 0:
            # Bayesian optimization
            search = BayesSearchCV(
                pipeline,
                param_space,
                n_iter=n_iter,
                cv=cv,
                scoring=scoring,
                refit='auprc',  # Optimize for AUPRC
                n_jobs=-1,
                random_state=42,
                verbose=0
            )
        else:
            # Random search fallback
            search = RandomizedSearchCV(
                pipeline,
                param_space,
                n_iter=n_iter,
                cv=cv,
                scoring=scoring,
                refit='auprc',  # Optimize for AUPRC
                n_jobs=-1,
                random_state=42,
                verbose=0
            )
        
        # Fit the search
        search.fit(X_train_processed, y_train)
        
        # Extract results
        best_idx = search.best_index_
        result = {
            'best_estimator': search.best_estimator_,
            'best_params': search.best_params_,
            'best_score_auprc': search.cv_results_['mean_test_auprc'][best_idx],
            'best_score_auprc_std': search.cv_results_['std_test_auprc'][best_idx],
            'best_score_roc_auc': search.cv_results_['mean_test_roc_auc'][best_idx],
            'best_score_roc_auc_std': search.cv_results_['std_test_roc_auc'][best_idx],
            'search_type': search_type,
            'n_iterations': n_iter
        }
        
        print(f"  âœ… AUPRC: {result['best_score_auprc']:.4f} (Â±{result['best_score_auprc_std']:.4f})")
        print(f"  âœ… ROC-AUC: {result['best_score_roc_auc']:.4f} (Â±{result['best_score_roc_auc_std']:.4f})")
        
        return result
        
    except Exception as e:
        print(f"  âŒ Failed: {str(e)[:100]}...")
        return None

print("âœ… Created optimized search functions:")
print("  â€¢ Bayesian optimization with AUPRC focus")
print("  â€¢ Random search fallback")
print("  â€¢ Reduced CV for faster iteration")
print("  â€¢ Comprehensive error handling")
print("")
```

```python
# ====================================================================
# MAIN OPTIMIZATION EXECUTION
# ====================================================================

print("ğŸš€ MAIN OPTIMIZATION EXECUTION")
print("="*35)
print("Running optimized ML pipeline with all improvements...")
print("")

# Results storage
results = {}
best_models = {}

# 1. BASELINE MODELS (NO SAMPLING)
print("ğŸ“Š 1. BASELINE MODELS (Built-in Imbalance Handling)")
print("-" * 50)

baseline_models = ['Balanced Random Forest', 'Easy Ensemble']
for model_name in baseline_models:
    pipeline = create_pipeline(model_name, sampler=None)
    
    # For balanced models, we don't need sampling parameters
    if model_name == 'Balanced Random Forest':
        param_space = get_random_forest_params(is_pipeline=False)
    elif model_name == 'Easy Ensemble':
        # EasyEnsemble has fewer tunable parameters
        param_space = {
            'n_estimators': Integer(5, 20),
            'sampling_strategy': Categorical(['auto'])
        }
    
    result = run_optimized_search(
        pipeline, param_space, f"Baseline_{model_name}", 
        search_type='bayes', n_iter=30
    )
    
    if result:
        results[f"Baseline_{model_name}"] = result
        best_models[f"Baseline_{model_name}"] = result['best_estimator']

print("\\n" + "="*60)

# 2. SAMPLING-BASED MODELS
print("ğŸ“Š 2. SAMPLING-BASED MODELS")
print("-" * 30)

# Test each sampling method with each compatible model
sampling_model_combinations = [
    ('ROS', ['Logistic Regression', 'Random Forest', 'XGBoost']),
    ('SMOTE', ['Logistic Regression', 'Random Forest', 'XGBoost']),
    ('KMeans-SMOTE', ['Logistic Regression', 'Random Forest', 'XGBoost']),
    ('SMOTE-ENN', ['Logistic Regression', 'Random Forest', 'XGBoost'])
]

for sampling_method, model_names in sampling_model_combinations:
    print(f"\\nğŸ¯ {sampling_method} Method:")
    print("-" * (len(sampling_method) + 10))
    
    sampler = sampling_methods[sampling_method]
    
    for model_name in model_names:
        pipeline_name = f"{sampling_method}_{model_name.replace(' ', '_')}"
        pipeline = create_pipeline(model_name, sampler)
        
        # Combine sampling and model parameters
        sampling_params = get_sampling_params(sampling_method)
        
        if model_name == 'Logistic Regression':
            model_params = get_logistic_params(is_pipeline=True)
        elif model_name == 'Random Forest':
            model_params = get_random_forest_params(is_pipeline=True)
        elif model_name == 'XGBoost':
            model_params = get_xgboost_params(is_pipeline=True)
        
        # Merge parameter spaces
        param_space = {**sampling_params, **model_params}
        
        result = run_optimized_search(
            pipeline, param_space, pipeline_name,
            search_type='bayes', n_iter=40
        )
        
        if result:
            results[pipeline_name] = result
            best_models[pipeline_name] = result['best_estimator']

print("\\n" + "="*60)
print(f"âœ… OPTIMIZATION COMPLETE! Tested {len(results)} model configurations")
print("")
```
